```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Dados - A análise dos dados foi feita pela área de inspeção de elemento no navegador ("DevTools"). 
# Foram contados os elementos da class="comment"> Comentário de exemplo' (nomes das pessoas), como também os class="comment"> Comentário de teste'(comentários anônimos). 

categorias = ["Condições de Trabalho", "Crescimento", "Bem-estar", "Autonomia"]
homens = [59, 59, 0, 0]
mulheres = [36, 36, 0, 0]
anonimos = [0, 0, 190, 95]

# Cores para as barras
cores = ['#ECB22E', '#E01E5A', '#2EB67E', '#36C5F0']

# Criar um DataFrame para os dados
data = pd.DataFrame({'Categorias': categorias,
                     'Homens': homens,
                     'Mulheres': mulheres,
                     'Anônimos': anonimos})

# Gráfico de barras empilhadas vertical com Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

bottom = [0] * len(categorias)
for genero, cor in zip(data.columns[1:], cores):
    plt.bar(data['Categorias'], data[genero], label=genero, color=cor, bottom=bottom)
    bottom = [i + j for i, j in zip(bottom, data[genero])]

plt.xlabel('Categorias')
plt.ylabel('Total de Comentários')
plt.title('Distribuição de Comentários por Categoria e Gênero')
plt.xticks(rotation=45)
plt.legend()

# Adicionar rótulos às barras
for index, row in data.iterrows():
    for genero in data.columns[1:]:
        total = row[genero]
        plt.text(index, total, str(total), ha='center', va='bottom')

plt.tight_layout()
plt.show()



```
